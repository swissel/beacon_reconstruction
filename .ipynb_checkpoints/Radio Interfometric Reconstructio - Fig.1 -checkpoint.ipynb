If you plot all three compoennts of the electric field, you'll see that the $E$-field (which is generated by the Lorentz force predominately) is mostly in the $y$-direction. (The antenna in the center, wi has a significant $E_z$ component). In principle, you should calculate the vector sum to get the total power (which is $\propto S_i(t)^2$).

The next step is to set up a 3D grid, where each point corresponds to the $j$-th position. Calculate the distance between each antenna 
and every point in the grid $d_{ij}$. 

That will tell you by how much 
to shift the waveforms:  $\Delta_{ij} = d_{ij}  n_{j} / c$. Evaluate the index of refraction
at the $j$-th position, point in the 3D grid you're evaluating. This should in principle be identical to the values in the sry file.

```python
#atmospheric refractive index
# 
a = 325*10**-6
b = 0.1218 #1/km
def n(z):
    # z corresponds to the height of the antenna array in km
    return 1+a*np.exp(-b*z)

height = np.linspace(0,10,11)
ARI = n(height)
plt.xlabel("Height (km)")
plt.ylabel("n-1")

plt.title("Atmospheric Refractive Index, $n$")
plt.plot(height, ARI-1., linestyle="-", color = "Blue")
```

Before going on, summing waveforms will be needed so this step shifts the Ey waveform so the maximum from each antenna is at the same time. Then it sums the componenets from all the antennas and plots all the waveforms. 

```python
temp_list = []
for nm in range(25):
    max_index_x = np.argmax(np.absolute(waveforms['Ey'][nm]))
    time_index_x = waveforms['t'][nm][max_index_x]
    temp_list.append(int(time_index_x))
max_time_offset_array = np.array(temp_list)
max_time_offset_array

#multiply by .5 because the t waveform array is going by 2 ns
s = np.zeros(len(waveforms["t"][0]))
for i in range(25):
    plt.plot(waveforms['t'][i], np.roll(waveforms['Ey'][i], int(-0.5*max_time_offset_array[i])), label=i+1)
    s+=np.roll(waveforms['Ey'][i], int(-0.5*max_time_offset_array[i]))
plt.xlim(-30,50)
#plt.ylim(-0.0002, 0.0002)
plt.title("$E_{y}$-Field on the Grid")
plt.xlabel("Time (ns)")
plt.ylabel("Ey")
#plt.legend()
plt.plot(waveforms['t'][i], s)
print(s[np.argmax(s)])
```

This step creates a 3D grid of points. Right now there is only one z axis coordinate which corresponds with the z coordinate where we want to take the slice later. 

```python
#creating grid of points
fig=plt.figure()
ax=fig.add_subplot(projection='3d')

#manually input size of the grid of points

xg = np.linspace(-200, 200, num=101)
        
yg = np.linspace(-200, 200, num=101)

zg = np.array([5250])

xx, yy, zz= np.meshgrid(xg, yg, zg,)


plt.title("Simulated grid positions")
ax.scatter(xx,yy,zz)

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')

plt.show()
```

```python
# this gets a point, keeps it xyz value on the grid, its xy value of the antenna, the seperation, and gets the delta_t value

columns = 7
rows = len(xg) * len(yg) * len(zg)
 
a_value = 325*10**-6
b_value = .0001218
c = 3*(10**8)

counter_1 = 0
antenna_counter = 0

list_of_arrays = []

#(x position, y position, z position, antenna x, antenna y, seperation, delta ij in ns)
for j in range(len(x)):
    for k in range(len(y)):
        antenna_counter+=1
        antenna_array = np.array([])
        for g in range(len(xg)):
            for h in range(len(yg)):
                for i in range(len(zg)):
                    temp_data_list = []
                    l = len(zg)
                    m = int(i%l)
                    n = 1 + a_value * np.exp((-1)*b_value * (zg[m]+1400))
                    delta_x = xg[g]-x[j]
                    delta_y = yg[h]-y[k]
                    delta_z = zg[i]
                    distance = np.sqrt(delta_x**2+delta_y**2+delta_z**2)
                    delta_tij = distance*n/c*(10**9)
                    
                    temp_data_list.append(xg[g])
                    temp_data_list.append(yg[h])
                    temp_data_list.append(zg[i])
                    temp_data_list.append(x[j])
                    temp_data_list.append(y[k])
                    temp_data_list.append(distance)
                    temp_data_list.append(delta_tij)
                    
                    temp_data_array_unshaped = np.array(temp_data_list)
                    temp_data_array_shaped = temp_data_array_unshaped.reshape(7,1)
                    
                    antenna_array = np.append(antenna_array, temp_data_list)
                    
                    counter_1+=1
        antenna_array = antenna_array.reshape(rows, columns)
        list_of_arrays.append(antenna_array)

array_of_arrays = np.array(list_of_arrays)
```

Here, for each point in the 3D grid made earlier, we calculate the light travel time to each antenna and shift the corrisponding waveform for each antenna. Then these waveforms get summed into a waveform called s and the max value of s is found. 

```python
s_max_list = []
for i in range(len(xg)*len(yg)):
    time_offset_list = []
    for k in range(25):
        time_offset_list.append(array_of_arrays[k][i][-1])
    time_offset_array = np.array(time_offset_list)
    s = np.zeros(len(waveforms["t"][0]))
    for k in range(25):
        s+=np.roll(waveforms['Ex'][k]+waveforms['Ey'][k]+waveforms['Ez'][k], int(-0.5*time_offset_array[k]))
    s_max = s[np.argmax(s)]
    one_s_max_array = [array_of_arrays[0][i][0], array_of_arrays[0][i][1], array_of_arrays[0][i][2], s_max]
    s_max_list.append(one_s_max_array)
s_max_array = np.array(s_max_list)
```

This step plots the max s value for each point in the xy plane. 

```python
plt.scatter(s_max_array[:,0], s_max_array[:,1], 5 ,c=s_max_array[:,3])
plt.title("Spacial Distribution of S")
plt.xlabel("x(m)")
plt.ylabel("y(m)")
cbar = plt.colorbar()
cbar.set_label("S value")
plt.show
```
